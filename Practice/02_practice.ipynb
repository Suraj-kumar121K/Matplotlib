# Bar Graph

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
student = np.array(["Suraj", "Mohan", "Amit", "Komal", "Shubham"])
marks = np.array([52, 78, 65, 42, 58])
```

```python
plt.bar(student, marks)

# Labels & Title
plt.xlabel("student")
plt.ylabel("marks")
plt.title("student marks")
plt.show()
```

### Step 1: Create Excel File with 50 Rows & 5 Columns

```python
# Random data generate kar rahe hain in numpy
first_names = ["Suraj", "Anish", "Riya", "Rahul", "Sneha", "Amit", "Neha", "Pooja", "Mohan", "Kiran"]
last_names = ["Kumar", "Sharma", "Singh", "Verma", "Patel", "Gupta", "Mehra", "Reddy", "Joshi", "Chauhan"]

data = {
    "Employee_ID": np.arange(1, 51),
    "Name": [f"{np.random.choice(first_names)} {np.random.choice(last_names)}" for _ in range(50)],
    "Age": np.random.randint(20, 60, 50),
    "Department": np.random.choice(["IT", "HR", "Sales", "Finance", "marketing"], 50),
    "Salary": np.random.randint(20000, 100000, 50)         
}

df = pd.DataFrame(data)
df.head(20)
```

```python
df.to_excel("Employees.xlsx", index=False)
```

Line Plot

```python
plt.figure(figsize=(12,8))
plt.bar(df["Name"], df["Salary"])
plt.xlabel("Name", fontweight="bold", fontsize=30)
plt.ylabel("salary", fontsize=30, color="y", fontweight="bold")
plt.title("Age Vs Salary", fontsize=30, color="r", fontweight="bold")

# X-axis labels ko tilt karo
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
dept_salary = df.groupby("Department")["Salary"].sum()
# Bar plot
plt.figure(figsize=(8,5))
plt.bar(dept_salary.index, dept_salary.values, color="skyblue")
plt.title("Sum Salary by Department")
plt.xlabel("Department")
plt.ylabel("Sum Salary")
plt.xticks(rotation=45)  # X-axis names thoda tilt kar do
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Style apply
plt.style.use('ggplot')

# Random Data
np.random.seed(2)
x1 = np.random.randint(1, 50, 15)
y1 = np.random.randint(20, 100, 15)

x2 = np.random.randint(1, 50, 15)
y2 = np.random.randint(20, 100, 15)

colors1 = np.random.rand(15)
colors2 = np.random.rand(15)

# Scatter Plot 1
plt.scatter(x1, y1,
            c=colors1,
            cmap='viridis',
            s=120,
            marker='o',
            edgecolor='black',
            alpha=0.8,
            label='Class A')

# Scatter Plot 2
plt.scatter(x2, y2,
            c=colors2,
            cmap='plasma',
            s=120,
            marker='^',
            edgecolor='black',
            alpha=0.8,
            label='Class B')

# Title & Labels
plt.title("Comparison of Class A and B Performance",
          fontsize=16,
          fontweight='bold')

plt.xlabel("Study Hours", fontsize=12)
plt.ylabel("Marks Obtained", fontsize=12)

# Axis Limits
plt.xlim(0, 60)
plt.ylim(0, 110)

# Grid
plt.grid(True, linestyle='--', linewidth=0.6)

# Legend
plt.legend()

plt.colorbar(label="Performance Intensity")

plt.show()
```

```python
# Data
t = np.linspace(0, 10, 200)
x = np.sin(t)
y = np.cos(t)
z = t

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Multiple 3D lines
ax.plot3D(x, y, z, color='blue', linewidth=2, label='Spiral 1')
ax.plot3D(y, x, z, color='red', linewidth=2, linestyle='--', label='Spiral 2')

# View angle control
ax.view_init(elev=30, azim=60)

# Axis limits
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(0, 10)

ax.legend()
plt.show()
```

```python
np.random.seed(1)

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(x, y, z,
                c=z,
                cmap='plasma',
                s=z*300,
                alpha=0.7)

fig.colorbar(sc)
plt.show()
```
