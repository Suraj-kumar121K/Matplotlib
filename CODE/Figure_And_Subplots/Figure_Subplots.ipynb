# Figure & Subplots in Matplotlib

###                          1Ô∏è‚É£ Figure
Figure is used to create a new figure (canvas) where one or more plots can be drawn.

##### üîπ Function
- Creates a new plotting area (figure).

##### üîπ Syntax
- `plt.figure()`

#### With size:
- `plt.figure(figsize=(width, height))`

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Single Plot Figure")
plt.show()
```

# 2Ô∏è‚É£ Subplot
Subplot is used to divide a figure into multiple rows and columns and place different plots in each section.

### üîπ Function
Creates a subplot inside a figure.

### üîπ Syntax
-`plt.subplot(nrows, ncols, index)`
- nrows ‚Üí number of rows
- ncols ‚Üí number of columns
- index ‚Üí position of the subplot (starts from 1)


```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Plot 1")
```

# 3Ô∏è‚É£ Multiple Plots in One Figure
Multiple plots can be displayed in one figure using `plt.subplot()`.

- This is useful for comparison and analysis.

üîπ Example

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Sales")

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [15, 25, 35])
plt.title("Profit")
plt.show()
```

# 4Ô∏è‚É£ plt.tight_layout()
`plt.tight_layout()` automatically adjusts spacing between subplots so that titles, labels, and plots do not overlap.

### üîπ Syntax
`plt.tight_layout()`

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [15, 25, 35])
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```
