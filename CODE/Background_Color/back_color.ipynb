# ✍️ Background Color in Matplotlib
The background color refers to the color behind the plotted area and the entire figure.

# ®️ Types of Background Color in Matplotlib
1. Figure Background Color
The figure background is the outer area of the plot that surrounds the graph.
- It is controlled using the facecolor parameter of the `figure()` function.

# Syntax:
 `plt.figure(facecolor='color_name')`

2. Axes Background Color
The axes background is the area where the actual data is plotted.
- It is also known as the plot area background.
- It can be changed using the set_facecolor() method.

# Syntax:
`ax.set_facecolor('color_name')`

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Create Dataset

```python
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9] 

# Create plots
fig = plt.Figure()
ax = plt.subplot()

ax.plot(a, b, 'k--', label='Frequency')
ax.plot(a, c, 'k:', label='Periods')

# Create a legend
l = ax.legend(loc = 'upper center')

# set the background color using the set_facecolor() function
l.get_frame().set_facecolor('red')

# plot title
plt.title("Frequency of a Signal")

# Display the figure
plt.show()
```
