# üéØ Plot Customization in Matplotlib (Very Important)
- plots clear, readable, and professional by adding titles, labels, legends, grid lines, and axis limits.

### 1Ô∏è‚É£ plt.title()
- `plt.title()` is used to add a title to a plot.
The title describes what the graph represents.

##### üîπ Syntax
- `plt.title("Title Name")`

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Sales Growth")
plt.show()
```

# 2Ô∏è‚É£ plt.xlabel() and plt.ylabel()
- `plt.xlabel()` sets the label for the X-axis
- `plt.ylabel()` sets the label for the Y-axis

Axis labels explain what data is shown on each axis.

## üîπ Syntax
- `plt.xlabel("X Axis Name")`
- `plt.ylabel("Y Axis Name")`

```python
plt.plot([1, 2, 3], [10, 20, 30])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
 
```

# 3Ô∏è‚É£ plt.legend()
`plt.legend()` is used to display labels for plotted data, especially when there are multiple lines or plots.

üîπ Syntax
`plt.legend()`

```python
plt.plot([1, 2, 3], [10, 20, 30], label="Product A")
plt.plot([1, 2, 3], [15, 25, 35], label="Product B")

plt.legend()
plt.show()
```

# 4Ô∏è‚É£ plt.grid()
plt.grid() adds grid lines to the plot, which helps in better reading and comparison of values.

üîπ Syntax
- `plt.grid(True)`

```python
plt.plot([1, 2, 3], [10, 20, 30])
plt.grid(True)
plt.show()
```

### Dashed Grid Lines (Style Example)
- `plt.grid(linestyle="--")`
Other styles:
- "-" ‚Üí solid
- "--" ‚Üí dashed
- ":" ‚Üí dotted
- "-." ‚Üí dash-dot

# 5Ô∏è‚É£ plt.xlim() and plt.ylim()
- `plt.xlim()` sets the range of the X-axis
- `plt.ylim()` sets the range of the Y-axis
- Used to control the visible area of the plot.

üîπ Syntax
- `plt.xlim(min, max)`
- `plt.ylim(min, max)`


```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40])

plt.xlim(1, 4)
plt.ylim(0, 50)

plt.show()
```

# ‚úÖ Complete Example (All Customizations Together)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y, label="Sales")

plt.title("Monthly Sales Report")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.xlim(1, 4)
plt.ylim(0, 50)

plt.show()
```
