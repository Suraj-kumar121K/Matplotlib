# ðŸŽ¨ Background Color in Matplotlib

### There are 2 type of Background color

1. Figure Background

2. Axes Background

# Figure Background Color
- Figure background poore graph ka outer area hota hai.

Iska color `plt.figure()` ke `facecolor` se change hota hai.

```python
import matplotlib.pyplot as plt

plt.figure(facecolor='red')
plt.plot([1, 2, 3], [10, 20, 30])
plt.show()
```

# 2: Axes Background Color
- Axes background graph ke andar ka part hota hai.
- Isko change karne ke liye `plt.gca()` use hota hai.

```python
plt.plot([1, 2, 3], [10, 20, 30])
plt.gca().set_facecolor('green')
plt.show()
```

# Figure + Axes dono ka color
- figure aur axes dono ka color alag hota hai.

```python
plt.figure(facecolor='lightblue')
plt.plot([1, 2, 3], [10, 20, 30])
plt.gca().set_facecolor('black')
plt.show()
```

# Object Oriented Method
- Advanced plotting me fig, ax = plt.subplots() use hota hai.

- jisse figure aur axes ko clearly control kar sakte hain.

```python
fig, ax = plt.subplots()

fig.patch.set_facecolor('lightgrey')  # figure background
ax.set_facecolor('lightgreen')        # axes background

ax.plot([1, 2, 3], [10, 20, 30])
plt.show()
```

# Transparency (Alpha)
Background ko thoda transparent banana ho
to `alpha` use karte hain.

```python
fig, ax = plt.subplots()
ax.set_facecolor('orange')
ax.patch.set_alpha(0.4)

ax.plot([1, 2, 3], [10, 20, 30])
plt.show()
```
