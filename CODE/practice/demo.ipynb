```python
import matplotlib.pyplot as plt
import numpy as np
x=[1, 2, 3, 4]
y=[2, 4, 6, 8]

c = ['r', 'y', 'g', 'b']
plt.bar(x, y, color=c)

plt. show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.title("Line Plot with Grid")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y, label="Sales")

plt.title("Monthly Sales Report")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.xlim(1, 4)
plt.ylim(0, 50)

plt.show()
```
