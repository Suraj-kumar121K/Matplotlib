# What is Legend in Graph
A legend is a box or label area in a graph that explains the meaning of different plot elements such as lines, markers, or colors.

# Why Use a Legend?
- In graphs with multiple data series, it is difficult to know which line or marker represents which series.
- The legend provides labels that make the graph self-explanatory.

# How to Add a Legend in Matplotlib
1. Assign a label to each plot element when plotting:
plt.plot(x, y, label='Series 1')
plt.plot(a, b, label='Series 2')

2. Call the legend() function to display the labels on the graph:
plt.legend()


```python
import matplotlib.pyplot as plt
import numpy as np
```

# Create Data

```python
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9] 
```

# Create Plots

```python
fig = plt.figure()
ax = plt.subplot()

ax.plot( a, b, 'k--', label='Frequency')
ax.plot(a, c, 'k:', label='Periods')

# Create a legend using the Matploting Axes. legend() Method
ax.legend()

# Plot Tital
plt.title("Frequecy of a Signal")

plt.show()
```

# ✔️ POSITION LEGENDS
We can easilt posistion legends in Matplotlib using the loc parameter of the legend() method. The following values set position:
- 'upper left', 'upper right', 'lower right'
- 'upper center', 'lower center', 'center right'
- 'center'

```python
fig = plt.figure()
ax = plt.subplot()

ax.plot( a, b, 'k--', label='Frequency')
ax.plot(a, c, 'k:', label='Periods')

# Create a legend using the Matploting Axes. legend() Method
ax.legend(loc='upper center')

# Plot Tital
plt.title("Frequecy of a Signal")

plt.show()
```
