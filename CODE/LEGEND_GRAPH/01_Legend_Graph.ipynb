# What is Legend in Graph
A legend is a box or label area in a graph that explains the meaning of different plot elements such as lines, markers, or colors.

# Why Use a Legend?
- In graphs with multiple data series, it is difficult to know which line or marker represents which series.
- The legend provides labels that make the graph self-explanatory.

# How to Add a Legend in Matplotlib
1. Assign a label to each plot element when plotting:
plt.plot(x, y, label='Series 1')
plt.plot(a, b, label='Series 2')

2. Call the legend() function to display the labels on the graph:
plt.legend()
