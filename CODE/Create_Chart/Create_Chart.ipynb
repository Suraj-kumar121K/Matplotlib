# ðŸ“Š Bar Graph in Matplotlib
- Bar graph ka use categorical data ko compare karne ke liye hota hai.
- Isme data rectangular bars ke form me dikhaya jata hai.

# ðŸ§© Syntax
`plt.bar(x, height)`
- x â†’ categories (names)
- height â†’ values (numbers)

# Example

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
subjects = ['Math', 'Science', 'English', "Hindi", "Phy"]
marks = [80, 70, 90, 55, 44]

# plot a bar graph
plt.bar(subjects, marks)

# The labels for x and y co-ordinaters
plt.xlabel("subjects")
plt.ylabel("marks")

# plot title
plt.title("Marks of Students")

# Display the figure
plt.show()
```
