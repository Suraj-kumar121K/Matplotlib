# Add GRID LINE
- Matplotlib me grid lines graph ko readable banane ke liye use hoti hain

4️⃣ Types
- Two type of grid line

1. Major Grid Lines: These lines represent the main scale of the axes
2. Minor Grid Lines: These lines show smaller divisions between major grid lines

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Cricket_Bar": ["SG", "BDM", "SS", "GM", "Kookaburra", "Spartan"],
    "MRP": [2000, 2200, 2400, 2700, 2800, 3000],
    "Weight_Grams": [1100, 1200, 1250, 1330, 1480, 1600]
}
```

```python
dataFrame = pd.DataFrame(data)
```

```python
plt.plot(dataFrame["MRP"], dataFrame["Weight_Grams"])
plt.grid()
plt.show()
```
