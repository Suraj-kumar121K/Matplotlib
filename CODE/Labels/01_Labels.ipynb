# WHAT IS LABELS
- Labels are used to describe the components of a plot such as the x-axis, y-axis, and the title.

# Types of Labels in Matplotlib
1. X-axis Label â€“ Describes values on the horizontal axis
2. Y-axis Label â€“ Describes values on the vertical axis
3. Title â€“ Provides a heading or description of the plot

# Functions Used for Labels
Labels are added using the following functions:
- xlabel() â†’ sets label for x-axis
- ylabel() â†’ sets label for y-axis
- title() â†’ sets the title of the plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Cricket_Bar": ["SG", "BDM", "SS", "GM", "Kookaburra", "Spartan"],
    "MRP": [2000, 2200, 2400, 2700, 2800, 3000],
    "Weight_Grams": [1100, 1200, 1250, 1330, 1480, 1600]
}
```

```python
dataFrame = pd.DataFrame(data)
```

# Titles and Position
A plot title is a descriptive heading shown at the top of a graph.

# ðŸ‘‹Positioning Plot Titles
Matplotlib allows changing the position of the plot title for better layout and readability.
The title position can be:

- Center (default position)
- Left aligned
- Right aligned

# loc in Title
- loc parameter is used to set the position (alignment) of the plot title.

# loc Parameter Values
The loc parameter supports the following positions:
- 'center' â†’ Title appears at the center (default)
- 'left' â†’ Title appears on the left side
- 'right' â†’ Title appears on the right side

```python
plt.plot(dataFrame["MRP"], dataFrame["Weight_Grams"])

# The labels for x and y co-ordinates
plt.xlabel("Bat Price (USA)")
plt.ylabel("Bat Weight (Grams)")

plt.title("Bat Price depends on the weight", loc="right") 

# Display the figure
plt.show()
```
