# üî§ Font Size in Matplotlib
Font size ka use graph ke text (title, labels, legend, ticks) ko bada ya chhota karne ke liye hota hai.

- Matplotlib me font size points (pt) me hota hai

# 1Ô∏è‚É£ Title ka Font Size
`plt.title("Sales Report", fontsize=16)`
- üìå Use: Graph ka title clearly visible banane ke liye

# 2Ô∏è‚É£ X-label & Y-label Font Size
`plt.xlabel("Month", fontsize=12)`
`plt.ylabel("Revenue", fontsize=12)`

# 3Ô∏è‚É£ Legend Font Size
`plt.legend(fontsize=10)`

# 5Ô∏è‚É£ Scatter / Text ka Font Size
`plt.text(2, 50, "Peak Value", fontsize=11)`
- üìå Use: Graph ke andar text likhne ke liye

# ü§î figure size
`plt.figure(figsize=(10, 5))`
üìå Meaning
- 10 ‚Üí width (inches)
- 5 ‚Üí height (inches)

# Example

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9] 

# Create plots
fig = plt.Figure()
ax = plt.subplot()

ax.plot(a, b, 'k--', label='Frequency')
ax.plot(a, c, 'k:', label='Periods')

# Create a legend
l = ax.legend(loc = 'upper center', fontsize='xx-large')

# plot title
plt.title("Frequency of a Signal")

# Display the figure
plt.show()
```
