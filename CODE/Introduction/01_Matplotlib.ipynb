# ðŸ“Š Matplotlib Plot Functions

# ðŸ”¹ bar() :- For a Bar Plot
- ðŸ‘‰ Categories ke beech comparison dikhata hai (vertical bars)

### Use:
- Marks comparison
- Sales comparison
- plt.bar(x, y)

# ðŸ”¹ hist() :- Plot a Histogram
- ðŸ‘‰ Data ka distribution / frequency dikhata hai

# Use:
- Range me values kitni baar aayi
- Continuous data
- plt.hist(data, bins=5)

# ðŸ”¹ pie() :- Plot a Pie Chart
- ðŸ‘‰ Data ka percentage / proportion dikhata hai

# Use:
- Market share
- Percentage data
- plt.pie(values, labels=labels)

# ðŸ”¹ scatter() :- Scatter Plot
- ðŸ‘‰ Do variables ke beech relationship show karta hai

# Use:
- Correlation dekhna
- Outliers find karna
- plt.scatter(x, y)

# ðŸ”¹ stem() :- Stem Plot
- ðŸ‘‰ Discrete data ko lines ke sath points me dikhata hai

# Use:
- Signal processing
- Discrete values
- plt.stem(x, y)

# ðŸ”¹ step() :- Step Plot
- ðŸ‘‰ Data ko step-wise (stairs) form me show karta hai

# Use:
- Sudden change dikhane ke liye
- Digital / sampled data
- plt.step(x, y)
